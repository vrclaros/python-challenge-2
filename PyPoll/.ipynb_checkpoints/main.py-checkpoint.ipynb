```python
#PyPoll

import os
import csv
csvpath = os.path.join('./', 'Resources','election_data.csv')

with open(csvpath) as csvfile:
    csvreader = csv.reader(csvfile, delimiter=',')
    
    #print(csvreader)
    csv_header = next(csvreader)
    #print(f"CSV Header: {csv_header}")
    #define variables and lists
    list_Candidate=[]
    #vote , county, candidate
    
    #Print Tiles
    lines = '-------------------------'
    print('Election Results')
    print(lines)

    #Total number of votes in dataset
    total_votes = 0

    for row in csvreader:
        #Total number of votes in dataset
        total_votes += 1
        
        #input values in list
        vote , county, candidate = row
        list_Candidate.append(candidate)        
    
    print("Total Votes:",total_votes)
    print(lines)
    
#define variables for Candidates
counter_Khan, counter_Correy, counter_Li, counter_Tooley = 0,0,0,0
candidate1, candidate2, candidate3, candidate4 = 'Khan', 'Correy', 'Li', "O'Tooley"

#create a Dictionary for results
dict_ElectionResults ={}

for row in list_Candidate:
    if row == (candidate1):
        counter_Khan += 1
    elif row == (candidate2):
        counter_Correy += 1
    elif row == (candidate3):
        counter_Li += 1
    elif row == (candidate4):
        counter_Tooley += 1

#Save results of Elections in dict
dict_ElectionResults = {candidate1:counter_Khan,candidate2:counter_Correy,
                       candidate3:counter_Li, candidate4:counter_Tooley}

#calculate Results
result1 = str(round((counter_Khan/total_votes)*100,4))
result2 = str(round((counter_Correy/total_votes)*100,4))
result3 = str(round((counter_Li/total_votes)*100,4))
result4 = str(round((counter_Tooley/total_votes)*100,4))

print(candidate1,":", result1 , "%" , "(",counter_Khan, ")")
print(candidate2,":", result2 , "%" , "(", counter_Correy, ")")
print(candidate3,":", result3, "%" , "(",counter_Li, ")")
print(candidate4,":", result4, "%" , "(",counter_Tooley, ")")
print(lines)

#Find Winner in ElectionResults and Print
winner_value = max(dict_ElectionResults.values())
winner_name = [k for k, v in dict_ElectionResults.items() if v == winner_value]
print('Winner: ', str(winner_name))
print(lines)

# Write output file .txt
# Specify the file to write to
output_path = os.path.join("./Resources/","Results_Poll.txt")

# Open the file using "write" mode. Specify the variable to hold the contents
with open(output_path, 'w') as output:
    
    output.write('Election Results\n')
    output.write(lines + '\n')
    output.write(candidate1 +': '+ result1 + "% " +"(" + str(counter_Khan) + ")\n")
    output.write(candidate2 +': '+ result2 + "% " +"(" + str(counter_Correy) + ")\n")
    output.write(candidate3 +': '+ result3 + "% " +"(" + str(counter_Li) + ")\n")
    output.write(candidate4 +': '+ result4 + "% " +"(" + str(counter_Tooley) + ")\n")
    output.write(lines + '\n')
    output.write('Winner: '+ str(winner_name) + '\n')
    output.write(lines + '\n')

```
